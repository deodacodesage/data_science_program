```python
month1_year1_sales = year1_sales[:4, 1:]
days = ["monday", "tuesday", "wednessday", "thursday", "friday", "saturday", "sunday"]
mean_sales = []

for i in range(month1_year1_sales.shape[1]):
    day_data = month1_year1_sales[:, i]
    mean_sales.append(day_data.mean())
    print (days[i], day_data, f"mean Sales ===> {day_data.mean()}")
```

```python
plt.bar(days, mean_sales)
```

```python
sales_copy = sales[1:, 1:] #sales_copy
month_averages = []

for i in range (0, len(sales_copy), 4):
    month_sales = sales_copy[i:i+4] #slice by 4week interval
    print(month_sales, "\n")
    month_mean = month_sales.mean()
    month_averages.append(month_mean)  #append new average to the list of averages

print(month_averages)
print()
highest_sales = max(month_averages) #get highest value in month list
print()
print(highest_sales)
print()
print(month_averages.index(highest_sales)) #get list postion of highest sales value
```

```python
sales_copy = sales[1:, 1:] #sales_copy
month_averages = []
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"] *3
for i, j in zip(list(range(0, len(sales_copy), 4)), list(range(27))):
    month_sales = sales_copy[i:i+4] #slice by 4week interval
    print (months[j])
    print(month_sales, "\n")
    month_mean = month_sales.mean()
    month_averages.append(month_mean)  #append new average to the list of averages

print(month_averages)
print()
highest_sales = max(month_averages) #get highest value in month list
print()
print(highest_sales)
print()
print(months[month_averages.index(highest_sales)]) #get month highest sales value
```
